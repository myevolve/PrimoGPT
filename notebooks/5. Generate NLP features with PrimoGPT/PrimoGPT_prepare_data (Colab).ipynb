```python
csv_file_name = f"{stock_symbol}_{start_date}_{end_date}_gpt.csv"

csv_file_path = os.path.join(data_dir, csv_file_name)

df = pd.read_csv(csv_file_path)

df.head(50)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np

def predict_movement(row, previous_direction):
    if row['News Relevance'] == 0 or (row['Sentiment'] + row['Trend Direction'] == 0):
        return previous_direction

    score = row['Sentiment'] + row['Trend Direction']

    if score > 0:
        return 'U'
    else:
        return 'D'

def classify_actual_movement(bin_label):
    if pd.isna(bin_label) or bin_label is None:
        return 'U'  # Pretpostavljamo 'Up' za nedostajuÄ‡e vrijednosti (ili moÅ¾ete odabrati 'D')
    bin_label = str(bin_label)  # Pretvaramo u string za svaki sluÄaj
    if bin_label.startswith('U'):
        return 'U'
    else:
        return 'D'

# Dodajemo stvarne vrijednosti za trenutni dan
df['Actual_Current_Day'] = df['Bin Label'].apply(classify_actual_movement)

# Inicijaliziramo listu za predviÄ‘anja
predictions = []

# Iteriramo kroz DataFrame i predviÄ‘amo za svaki red
previous_direction = 'U'  # PoÄetna pretpostavka za prvi red
for _, row in df.iterrows():
    prediction = predict_movement(row, previous_direction)
    predictions.append(prediction)
    previous_direction = row['Actual_Current_Day']  # AÅ¾uriramo previous_direction za sljedeÄ‡u iteraciju

# Dodajemo predviÄ‘anja u DataFrame
df['Predicted_Next_Day'] = predictions

# PomiÄemo stvarne vrijednosti za jedan dan unaprijed za usporedbu s predviÄ‘anjima
df['Actual_Next_Day'] = df['Actual_Current_Day'].shift(-1)

# Uklanjamo zadnji red jer nemamo stvarnu vrijednost za sljedeÄ‡i dan nakon zadnjeg dana
df = df.dropna(subset=['Actual_Next_Day'])

# RaÄunamo toÄnost
accuracy = accuracy_score(df['Actual_Next_Day'], df['Predicted_Next_Day'])
conf_matrix = confusion_matrix(df['Actual_Next_Day'], df['Predicted_Next_Day'], labels=['U', 'D'])

print(f"ToÄnost predviÄ‘anja: {accuracy:.2f}")
print("\nMatrica zabune:")
print(pd.DataFrame(conf_matrix, index=['Actual U', 'Actual D'], columns=['Pred U', 'Pred D']))

print("\nIzvjeÅ¡taj o klasifikaciji:")
print(classification_report(df['Actual_Next_Day'], df['Predicted_Next_Day']))


# Analiza toÄnosti kombinacije sentimenta i trenda
correct_predictions = ((df['Sentiment'] + df['Trend Direction'] > 0) & (df['Actual_Next_Day'] == 'U')) | \
                      ((df['Sentiment'] + df['Trend Direction'] < 0) & (df['Actual_Next_Day'] == 'D')) | \
                      ((df['Sentiment'] + df['Trend Direction'] == 0) & (df['Predicted_Next_Day'] == df['Actual_Next_Day']))
combined_accuracy = correct_predictions.mean()
print(f"\nToÄnost kombinacije Sentiment + Trend Direction: {combined_accuracy:.2f}")
```

```python
def improved_ensemble_predict(row, previous_direction):
    predictions = []
    weights = []

    # Predikcija 1: Bazirano na Trend Direction (najbolji pojedinaÄni prediktor)
    if row['Trend Direction'] > 0:
        predictions.append('U')
        weights.append(3)  # Dajemo veÄ‡u teÅ¾inu ovom prediktoru
    elif row['Trend Direction'] < 0:
        predictions.append('D')
        weights.append(3)
    else:
        predictions.append(previous_direction)
        weights.append(1)

    # Predikcija 2: Bazirano na Investor Confidence i Price Impact Potential
    score2 = row['Investor Confidence'] + row['Price Impact Potential']
    if score2 > 0:
        predictions.append('U')
        weights.append(2)  # Srednja teÅ¾ina
    elif score2 < 0:
        predictions.append('D')
        weights.append(2)
    else:
        predictions.append(previous_direction)
        weights.append(1)

    # Predikcija 3: Bazirano na Sentiment
    if row['Sentiment'] > 0:
        predictions.append('U')
        weights.append(1)  # Manja teÅ¾ina
    elif row['Sentiment'] < 0:
        predictions.append('D')
        weights.append(1)
    else:
        predictions.append(previous_direction)
        weights.append(1)

    # Predikcija 4: Bazirano na Risk Profile Change (dobar za predviÄ‘anje D trendova)
    if row['Risk Profile Change'] > 0:
        predictions.append('D')  # Obrnuta logika jer veÄ‡i rizik sugerira pad
        weights.append(2)
    elif row['Risk Profile Change'] < 0:
        predictions.append('U')
        weights.append(1)  # Manja teÅ¾ina za U predikciju
    else:
        predictions.append(previous_direction)
        weights.append(1)

    # Predikcija 5: Bazirano na Earnings Impact
    if row['Earnings Impact'] > 0:
        predictions.append('U')
        weights.append(1)  # Manja teÅ¾ina
    elif row['Earnings Impact'] < 0:
        predictions.append('D')
        weights.append(1)
    else:
        predictions.append(previous_direction)
        weights.append(1)

    # VraÄ‡amo teÅ¾insku veÄ‡inu
    u_weight = sum(w for p, w in zip(predictions, weights) if p == 'U')
    d_weight = sum(w for p, w in zip(predictions, weights) if p == 'D')

    return 'U' if u_weight >= d_weight else 'D'

# Inicijaliziramo listu za improved ensemble predviÄ‘anja
improved_ensemble_predictions = []

# Iteriramo kroz DataFrame i predviÄ‘amo za svaki red
previous_direction = 'U'  # PoÄetna pretpostavka za prvi red
for _, row in df.iterrows():
    prediction = improved_ensemble_predict(row, previous_direction)
    improved_ensemble_predictions.append(prediction)
    previous_direction = row['Actual_Current_Day']  # AÅ¾uriramo previous_direction za sljedeÄ‡u iteraciju

# Dodajemo improved ensemble predviÄ‘anja u DataFrame
df['Improved_Ensemble_Prediction'] = improved_ensemble_predictions

# IzraÄun toÄnosti improved ensemble modela
improved_ensemble_accuracy = accuracy_score(df['Actual_Next_Day'], df['Improved_Ensemble_Prediction'])
print(f"ToÄnost improved ensemble modela: {improved_ensemble_accuracy:.2f}")

print("\nMatrica zabune improved ensemble modela:")
improved_ensemble_conf_matrix = confusion_matrix(df['Actual_Next_Day'], df['Improved_Ensemble_Prediction'], labels=['U', 'D'])
print(pd.DataFrame(improved_ensemble_conf_matrix, index=['Actual U', 'Actual D'], columns=['Pred U', 'Pred D']))

print("\nDetaljna analiza improved ensemble modela:")
print(classification_report(df['Actual_Next_Day'], df['Improved_Ensemble_Prediction']))
```

```python
def analyze_feature_predictive_power(df, features):
    results = {}

    for feature in features:
        predictions = []
        previous_direction = 'U'  # PoÄetna pretpostavka za prvi red

        for _, row in df.iterrows():
            if row[feature] > 0:
                prediction = 'U'
            elif row[feature] < 0:
                prediction = 'D'
            else:
                prediction = previous_direction

            predictions.append(prediction)
            previous_direction = row['Actual_Current_Day']

        accuracy = accuracy_score(df['Actual_Next_Day'], predictions)
        conf_matrix = confusion_matrix(df['Actual_Next_Day'], predictions, labels=['U', 'D'])
        report = classification_report(df['Actual_Next_Day'], predictions, output_dict=True)

        results[feature] = {
            'accuracy': accuracy,
            'confusion_matrix': conf_matrix,
            'classification_report': report
        }

    return results

# Lista znaÄajki koje Å¾elimo analizirati
features_to_analyze = [
    "Sentiment", "Price Impact Potential",
    "Trend Direction", "Earnings Impact", "Investor Confidence", "Risk Profile Change"
]

# Provodimo analizu
feature_analysis = analyze_feature_predictive_power(df, features_to_analyze)

# Ispisujemo rezultate
for feature, results in feature_analysis.items():
    print(f"\nAnaliza za znaÄajku: {feature}")
    print(f"ToÄnost: {results['accuracy']:.2f}")
    print("\nMatrica zabune:")
    print(pd.DataFrame(results['confusion_matrix'], index=['Actual U', 'Actual D'], columns=['Pred U', 'Pred D']))
    print("\nDetaljna analiza:")
    print(f"Precision (U): {results['classification_report']['U']['precision']:.2f}")
    print(f"Recall (U): {results['classification_report']['U']['recall']:.2f}")
    print(f"F1-score (U): {results['classification_report']['U']['f1-score']:.2f}")
    print(f"Precision (D): {results['classification_report']['D']['precision']:.2f}")
    print(f"Recall (D): {results['classification_report']['D']['recall']:.2f}")
    print(f"F1-score (D): {results['classification_report']['D']['f1-score']:.2f}")
    print("-" * 50)

# Sortiramo znaÄajke prema toÄnosti
sorted_features = sorted(feature_analysis.items(), key=lambda x: x[1]['accuracy'], reverse=True)

print("\nZnaÄajke rangirane prema toÄnosti:")
for feature, results in sorted_features:
    print(f"{feature}: {results['accuracy']:.2f}")
```
